```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# ##  Dataset:
# | Size (sqft) | Number of Bedrooms  | Number of floors | Age of  Home | Price (1000s dollars)  |
# | ----------------| ------------------- |----------------- |--------------|----------------------- |
# | 952             | 2                   | 1                | 65           | 271.5                  |
# | 1244            | 3                   | 2                | 64           | 232                    |
# | 1947            | 3                   | 2                | 17           | 509.8                  |
# | ...             | ...                 | ...              | ...          | ...                    |
```

```python
x_train=np.array([[952.0,2.0,1.0,65.0],[1244.0,3.0,2.0,64.0],[1947.0,3.0,2.0,17.0]])
y_train=np.array([271.5,332.0,509.8])
# again same gradient descnt :
alpha = 0.00001   # tiny learning rate
epoch=1000
costs = []
iters=[]
w=np.array([1.0,1.0,1.0,1.0])
b=1.0
m=len(x_train)
for i in range(epoch):
  hypo= np.dot(x_train,w) + b
  cost = (1/(2*m)) * sum(((hypo- y_train)**2))
  dw= (1/m) * np.dot(x_train.T,hypo-y_train)
  db=(1/m)* sum((hypo-y_train ))
  temp_w=w- alpha* dw
  temp_b=b- alpha*db
  w=temp_w
  b=temp_b
  if(i%10==0):
    costs.append(cost)
    iters.append(i)


plt.plot(range(0, epoch, 10), costs)
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.title("Cost vs Iterations")
plt.show()


```
